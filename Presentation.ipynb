## Interactive output

````python
from ipywidgets import interact

@interact(a=(0.1, 5.0), func=['sin', 'cos', 'tan'])
def plot_func(a=1, func='sin'):
    x = np.linspace(0.01, 2*np.pi, 200)
    y = getattr(np, func)(a/x)
    plt.plot(x, y)
````
